```python
#parameters
theta_12 = math.asin(np.sqrt(0.307))  
dm_21 = np.sqrt(7.53e-5)
theta_23 = math.asin(np.sqrt(0.536))
dm_32 = np.sqrt(0.00255)
theta_13 = math.asin(np.sqrt(0.0218))
```

```python
L = np.linspace(0,100000, 1000)
L;
```

```python
#values of the energy
E_1 = 5
E_2 = 2

#other parameters
dm_31 = dm_32+dm_21
alpha= dm_21**2/dm_31**2
# A = 2*E*V/dm_31**2
```

```python
delta_1 = ((dm_31**2) *L)/4*E_1
delta_2 = ((dm_31**2) *L)/4*E_2
```

```python
#other parameters
c_13 = np.sqrt(1-(np.sin(theta_13))**2)
c_12 = np.sqrt(1-(np.sin(theta_12))**2)
```

```python
term_1 = ((c_13**4) *((np.sin(2*theta_12))**2) * (np.sin(alpha*delta_1))**2)
term_10 = ((c_13**4) *((np.sin(2*theta_12))**2) * (np.sin(alpha*delta_2))**2)
```

```python
term_2 = (((np.sin(2*theta_13))**2) *(np.sin((alpha-1)*delta_1))**2)
term_20= (((np.sin(2*theta_13))**2) *(np.sin((alpha-1)*delta_2))**2)
```

```python
term_3 = (0.5*(c_12**2) *(np.sin(2*theta_13))**2 *(np.cos(2*delta_1) - np.cos(2*(alpha-1)*delta_1)))
term_30 = (0.5*(c_12**2) *(np.sin(2*theta_13))**2 *(np.cos(2*delta_2) - np.cos(2*(alpha-1)*delta_2)))
```

```python
P_O = 1-term_1-term_2+term_3
P_OO = 1- term_10 -term_20 +term_30
```
